```python

s3 = boto3.client('s3')


paginator = s3.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=raw_data_key)

folder_files = defaultdict(list)

for page in pages:
    for obj in page.get('Contents', []):
        key = obj['Key']
        if '/' in key:
            folder = key.split('/')[0]
            folder_files[folder].append(key)
            
folder = "musan/noise/free-sound"
noise_files = [f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f))]


# Now loop through each folder and its two files
for folder, files in folder_files.items():
    
    # Get transcript
    audio_file = files[0]
    transcript_file = files[1]
    base = "converted_audio/" + os.path.splitext(audio_file)[0] + ".wav"
    
    audio = s3.get_object(Bucket=raw_data_key, Key=audio_file)
    transcript = s3.get_object(Bucket=raw_data_key, Key=transcript_file)
    
    with open(transcript, "r", encoding="utf-8") as f:
        transcript_text = f.read()
    
    # Convert audio format
    
    ffmpeg.input(audio).output(base, ar=16000, ac=1, acodec='pcm_s16le').run()
    
    clean_audio = AudioSegment.from_file(base)
    
    # Noisify some of the audio files with random noise options (Help vary the mic quality etc.)
    
    noise_file = random.choice(noise_files)
    noise_audio = AudioSegment.from_file(folder + noise_file)
    
    if len(noise_audio) < len(clean_audio):
        noise_audio = noise_audio * (len(clean_audio) // len(noise_audio) + 1)
    noise_audio = noise_audio[:len(clean_audio)]
    
    noise_db_range = (-15, -5)
    noise_db = random.randint(*noise_db_range)
    noise_audio = noise_audio + noise_db
    
    start = random.randint(0, len(noise_audio) - len(clean_audio))
    noise = noise_audio[start:start + len(clean_audio)]

    # Mix
    combined = clean_audio.overlay(noise)
    combined.export(base, format="wav")
    
    
    # Trim silence
    audio = AudioSegment.from_file(base + "wav")
    trimmed = silence.strip_silence(audio, silence_thresh=-40, padding=500)
    trimmed.export(base + "wav", format="wav")
    
    # Break audio/transcription into 30 second sound bytes (Whisper works best with this or smaller audio set)
    # Or don't and say we did
    
    
    # Move into processed data bucket
    s3.upload_file(base + "wav", output_data_uri, base + "wav")
    
```
